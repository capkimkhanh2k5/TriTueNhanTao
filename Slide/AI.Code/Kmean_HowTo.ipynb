```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Countries-exercise.csv')
```

```python
import os
print(os.getcwd())
!ls
```

```python
X = data.iloc[:, 1:3]
```

```python
X
```

```python
plt.scatter(X['Longitude'], X['Latitude'])
```

```python
# Clustering
K = 3
kmeans = KMeans(K)
kmeans.fit(X)
```

```python
# Clustering Resutls
clusters = kmeans.predict(X)
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = clusters
```

```python
data_with_clusters
```

```python
#Plot the Resutls
plt.scatter(data_with_clusters['Longitude'],
            data_with_clusters['Latitude'],
            c = data_with_clusters['Clusters'],
            cmap ='rainbow')
```

# How to choose K
1. Tính tổng bình phương khoảng cách đến mỗi tâm
2. kmeans.inertia_ hỗ trợ làm việc đó
3. K = 1, thì kmeans.inertia_ là lớn nhất

# Mục mới

```python
l = []
for i in range(1,10):
    kmns = KMeans(i)
    kmns.fit(X)
    l_iter = kmns.inertia_
    l.append(l_iter)
```

```python
l
```

```python
plt.plot(range(1,10), l)
```

```python
# K = 5 phù hợp hơn :)
K = 5
kmeans = KMeans(K)
kmeans.fit(X)
```

```python
clusters = kmeans.predict(X)
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['Longitude'],
            data_with_clusters['Latitude'],
            c = data_with_clusters['Clusters'],
            cmap ='rainbow')
```
